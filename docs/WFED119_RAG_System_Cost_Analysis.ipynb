### 3.2 Data Source Transformation Strategy

#### Source 1: conversation_messages

| Step | Action | Rationale |
|------|--------|----------|
| **Group** | Combine messages by session_id | Individual messages lack context |
| **Filter** | Keep sessions with 4+ exchanges | Short sessions have low value |
| **Summarize** | AI-generate session summary | Better embedding quality |
| **Extract** | Identify strengths/values mentioned | Structured metadata |
| **Anonymize** | Remove names, emails, phone numbers | Privacy protection |

**Embedding Text Structure:**
```
## Career Coaching Session Summary
Stage: exploration
Topic: Project management experience

### Key Discussion Points:
- Led cross-functional team of 8 people
- Delivered project 2 weeks early
- Enjoyed coordination aspects

### Strengths Identified:
- Team Leadership
- Communication
- Project Coordination
```

#### Source 2: strength_profiles

| Step | Action | Rationale |
|------|--------|----------|
| **Normalize** | Unify synonyms ("leadership" = "team leadership") | Improve search accuracy |
| **Classify** | Assign user type based on skill patterns | Enable pattern matching |
| **Generate** | Create profile narrative | Rich embedding context |

**Embedding Text Structure:**
```
## Career Strength Profile

### Core Skills:
- Problem-Solving
- Communication
- Data Analysis

### Work Attitudes:
- Detail-oriented
- Collaborative

### Profile Summary:
This individual excels in analytical roles requiring
strong communication and collaborative problem-solving.
```

#### Source 3: value_results

| Step | Action | Rationale |
|------|--------|----------|
| **Combine** | Merge terminal + instrumental + work values | Holistic view per user |
| **Interpret** | Generate value profile narrative | Career alignment context |
| **Suggest** | Add career alignment hints | Actionable recommendations |

**Embedding Text Structure:**
```
## Personal Values Profile

### Life Goals (Terminal Values):
1. Achievement

### Career Priorities (Work Values):
1. Independence

3. Recognition

```

#### Source 4: LifeCraft Textbook (PDF)

| Step | Action | Rationale |
|------|--------|----------|
| **Extract** | OCR/parse PDF content | Get text from PDF |
| **Chunk** | Split into ~500 token segments | Optimal embedding size |
| **Overlap** | 50 token overlap between chunks | Preserve context boundaries |
| **Tag** | Add chapter/page metadata | Enable source citation |

### 3.3 Embedding Storage Schema

All processed data is stored in a unified embedding table:

| Column | Type | Description |
|--------|------|-------------|
| `id` | UUID | Primary key |
| `source_type` | TEXT | 'conversation', 'strength', 'value', 'textbook' |
| `source_id` | TEXT | Reference to original record |
| `user_id` | UUID | Owner (null for textbook) |
| `content` | TEXT | Structured text for embedding |
| `embedding` | VECTOR(1536) | OpenAI embedding vector |
| `metadata` | JSONB | Filtering tags (stage, topic, etc.) |
| `is_anonymized` | BOOLEAN | True if contains anonymized user data |

### 3.4 Embedding Update Strategy

| Data Type | Update Frequency | Trigger |
|-----------|------------------|----------|
| Textbook | One-time | Initial setup |
| Conversations | Daily batch | Cron job (2 AM) |
| Strength profiles | On creation | After session completion |
| Value results | On update | After assessment completion |

## 4. Cost Analysis: 4-Year Projection

### 4.1 User Growth Assumptions

| Year | New Users | Cumulative | Active Sessions/Year |
|------|-----------|------------|----------------------|
| Year 1 | 1,000 | 1,000 | 8,000 |
| Year 2 | 2,000 | 3,000 | 20,000 |
| Year 3 | 4,000 | 7,000 | 40,000 |
| Year 4 | 3,000 | 10,000 | 50,000 |
| **Total** | **10,000** | - | **118,000** |

**Assumptions:**
- 16 sessions/user/year (8 per semester Ã— 2 semesters)
- ~50% of cumulative users remain active
- 8 messages per session average

````python
# =============================================================================
# WFED119 RAG Cost Calculator
# =============================================================================

# Model Pricing (November 2025)
HAIKU_INPUT = 0.80 / 1_000_000    # $0.80 per 1M input tokens
HAIKU_OUTPUT = 4.00 / 1_000_000   # $4.00 per 1M output tokens
SONNET_INPUT = 3.00 / 1_000_000   # $3.00 per 1M input tokens
SONNET_OUTPUT = 15.00 / 1_000_000 # $15.00 per 1M output tokens
EMBEDDING = 0.02 / 1_000_000      # $0.02 per 1M tokens

# Token usage per session (8 messages)
WITHOUT_RAG = {'input': 12000, 'output': 3200}
WITH_RAG = {'input': 6000, 'output': 3200, 'embed': 2000}  # 50% input reduction

# Session projections
SESSIONS = {'Year 1': 8000, 'Year 2': 20000, 'Year 3': 40000, 'Year 4': 50000}

# Infrastructure costs per year
INFRA = {'Year 1': 720, 'Year 2': 900, 'Year 3': 1200, 'Year 4': 1500}

def calc_ai_cost(sessions, with_rag=True):
    if with_rag:
        haiku = sessions * 0.8  # 80% Haiku
        sonnet = sessions * 0.2  # 20% Sonnet
        return (
            haiku * WITH_RAG['input'] * HAIKU_INPUT +
            haiku * WITH_RAG['output'] * HAIKU_OUTPUT +
            sonnet * WITH_RAG['input'] * SONNET_INPUT +
            sonnet * WITH_RAG['output'] * SONNET_OUTPUT +
            sessions * WITH_RAG['embed'] * EMBEDDING
        )
    else:
        return (
            sessions * WITHOUT_RAG['input'] * HAIKU_INPUT +
            sessions * WITHOUT_RAG['output'] * HAIKU_OUTPUT
        )

# Calculate totals
print("=" * 65)
print("WFED119 4-Year Cost Projection (10,000 Users)")
print("=" * 65)

total_rag = 0
total_no_rag = 0

for year, sess in SESSIONS.items():
    ai_rag = calc_ai_cost(sess, True)
    ai_no_rag = calc_ai_cost(sess, False)
    infra = INFRA[year]
    
    total_rag += ai_rag + infra
    total_no_rag += ai_no_rag + infra
    
    print(f"\n{year} ({sess:,} sessions):")
    print(f"  AI (with RAG):    ${ai_rag:>8,.2f}")
    print(f"  AI (without RAG): ${ai_no_rag:>8,.2f}")
    print(f"  Infrastructure:   ${infra:>8,.2f}")
    print(f"  Savings (RAG):    ${ai_no_rag - ai_rag:>8,.2f} ({(ai_no_rag-ai_rag)/ai_no_rag*100:.0f}%)")

# Add one-time RAG implementation cost
RAG_SETUP = 2000
total_rag += RAG_SETUP

print("\n" + "=" * 65)
print("4-YEAR SUMMARY")
print("=" * 65)
print(f"\nWith RAG System:")
print(f"  AI + Infrastructure:  ${total_rag - RAG_SETUP:>10,.2f}")
print(f"  RAG Setup (one-time): ${RAG_SETUP:>10,.2f}")
print(f"  Contingency (20%):    ${(total_rag)*0.2:>10,.2f}")
print(f"  â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
print(f"  TOTAL:                ${total_rag * 1.2:>10,.2f}")
print(f"  Cost per user:        ${total_rag * 1.2 / 10000:>10,.2f}")

print(f"\nWithout RAG System:")
print(f"  AI + Infrastructure:  ${total_no_rag:>10,.2f}")
print(f"  Contingency (20%):    ${total_no_rag*0.2:>10,.2f}")
print(f"  â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
print(f"  TOTAL:                ${total_no_rag * 1.2:>10,.2f}")
print(f"  Cost per user:        ${total_no_rag * 1.2 / 10000:>10,.2f}")

print(f"\nğŸ’° 4-Year Savings with RAG: ${(total_no_rag - total_rag) * 1.2:,.2f}")
````

## 6. Comparison with ScholaRAG Project

| Aspect | ScholaRAG | WFED119 |
|--------|-----------|----------|
| **Purpose** | Systematic Literature Review | Career Coaching Chatbot |
| **Target Users** | 60 researchers | 10,000 students |
| **Primary Model** | Haiku 60% / Sonnet 40% | Haiku 80% / Sonnet 20% |
| **RAG Sources** | 15,000 papers/user | Textbook + user data |
| **Token Volume** | Very High (paper screening) | Medium (conversations) |
| **Cost/User/Year** | ~$70 | ~$2.50 |
| **4-Year Budget** | $48,000 (pilot: $12K) | **$25,000** |

### Why WFED119 is More Cost-Efficient

1. **Shorter interactions**: 8 messages vs 100s of paper screens
2. **Higher Haiku usage**: Simpler conversational tasks
3. **Smaller RAG corpus**: One textbook vs thousands of papers
4. **Shared knowledge base**: Textbook embeddings reused across all users
